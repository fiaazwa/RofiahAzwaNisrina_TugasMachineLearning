```python
df = pd.read_csv('/content/drive/MyDrive/Rofiah/TM9/Dataset/diabetes.csv')
#check data has ben read in properly
df.head()
```

```python
#check number of rows and coloms in dataset
df.shape
```

```python
#create a dataframe with a;; training data except the target colomn
X = df.drop(columns=['Outcome'])

#check that the target variable has been removed
X.head()
```

```python
#separate target values
y = df['Outcome'].values

#view target values
y[0:10]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1, stratify=y)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3)

knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
from sklearn import metrics

print ("hasil klasifikasi : \n", metrics.classification_report(y_test, y_pred))
print ("confusion metrics : \n", metrics.confusion_matrix(y_test, y_pred))
```
